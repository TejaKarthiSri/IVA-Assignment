```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # For displaying images in Colab
from google.colab import files

# Upload image
uploaded = files.upload()

# Load the image
image = cv2.imread(list(uploaded.keys())[0])

# Step 1: Sobel Edge Detection
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)  # Horizontal gradient
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)  # Vertical gradient

# Combine Sobel X and Y
sobel_combined = cv2.magnitude(sobel_x, sobel_y)
sobel_combined = cv2.convertScaleAbs(sobel_combined)

# Step 2: Canny Edge Detection
canny_edges = cv2.Canny(image, 100, 200)  # Threshold values can be adjusted

# Display Results
print("Original Image")
cv2_imshow(image)

print("Sobel X")
cv2_imshow(sobel_x)

print("Sobel Y")
cv2_imshow(sobel_y)

print("Combined Sobel")
cv2_imshow(sobel_combined)

print("Canny Edge Detection")
cv2_imshow(canny_edges)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # For displaying images in Colab
from google.colab import files

uploaded = files.upload()

# Load the original image
image = cv2.imread(list(uploaded.keys())[0])

# Function to compute T-Pyramid
def compute_t_pyramid(image, levels=4):
    pyramid = [image]  # Start with the original image

    # Generate down-sampled images for the pyramid
    for i in range(1, levels):
        image = cv2.pyrDown(image)  # Down-sample the image
        pyramid.append(image)

    # Display all pyramid levels
    for i, img in enumerate(pyramid):
        print(f"Level {i}: {img.shape}")
        cv2_imshow(img)

# Compute and display the T-Pyramid
compute_t_pyramid(image)
```

```python
import cv2
from google.colab.patches import cv2_imshow
from google.colab import files

# Upload image
uploaded = files.upload()

# Load the image
image = cv2.imread(list(uploaded.keys())[0])

# Apply Gaussian Blur
blurred_image = cv2.GaussianBlur(image, (15, 15), 0)

# Display results
cv2_imshow(image)
cv2_imshow(blurred_image)
```
